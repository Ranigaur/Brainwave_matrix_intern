```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
df=pd.read_csv("Supermart Grocery Sales - Retail Analytics Dataset.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'],format='mixed', dayfirst=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
# 4. Total sales and profit summary
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
print(f"\n✅ Total Sales: ₹{total_sales:,.2f}")
print(f"✅ Total Profit: ₹{total_profit:,.2f}")

```

```python


sales_trend = df.groupby('Order Date')['Sales'].sum().reset_index()

fig = px.line(
    sales_trend,
    x='Order Date',
    y='Sales',
    title='📈 Daily Sales Trend',
    labels={'Sales': 'Sales (₹)', 'Order Date': 'Date'},
    color_discrete_sequence=['teal']
)

fig.update_layout(xaxis_tickangle=45)
fig.show()
```

```python


top_categories = df.groupby('Category')['Sales'].sum().sort_values().reset_index()

fig = px.bar(
    top_categories,
    x='Sales',
    y='Category',
    orientation='h',
    color='Sales',
    color_continuous_scale='Oranges',
    title='📊 Sales by Category',
    labels={'Sales': 'Sales (₹)', 'Category': 'Category'}
)

fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python


df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
monthly_sales['Month'] = monthly_sales['Month'].astype(str)

fig = px.line(
    monthly_sales,
    x='Month',
    y='Sales',
    title="🗓️ Monthly Sales Trend",
    markers=True,
    labels={'Sales': 'Sales (₹)', 'Month': 'Month'}
)

fig.update_layout(xaxis_tickangle=45)
fig.show()
```

```python


top_sub = df.groupby('Sub Category')['Profit'].sum().sort_values(ascending=False).head(10)

fig = px.bar(
    x=top_sub.values,
    y=top_sub.index,
    orientation='h',
    color=top_sub.values,
    color_continuous_scale='Greens_r',
    labels={'x': 'Profit (₹)', 'y': 'Sub Category'},
    title='💰 Top 10 Profitable Sub-Categories'
)

fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python


# Group data
region_sales = df.groupby('Region')['Sales'].sum().reset_index()

# Create pie chart
fig = px.pie(
    region_sales,
    names='Region',
    values='Sales',
    title='📍 Sales Distribution by Region',
    hole=0.3  # Optional: donut-style chart
)

fig.update_traces(textinfo='percent+label')
fig.write_html("interactive_region_sales_pie.html")  # Save as HTML
fig.show()
```

```python


# Ensure your data (df) is already loaded and clean

fig = px.scatter(
    df,
    x='Discount',
    y='Profit',
    color='Category',  # Same as 'hue' in seaborn
    title='💸 Discount vs Profit',
    hover_data=['Sub Category', 'Sales', 'Customer Name'],  # Optional extra info on hover
    size_max=10
)

fig.update_layout(xaxis_title='Discount (%)', yaxis_title='Profit (₹)')
fig.write_html("interactive_discount_vs_profit.html")
fig.show()
```
